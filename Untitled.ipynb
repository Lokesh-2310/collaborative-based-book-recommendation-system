### Popularity based recommendation system

```python
ratings_with_name = ratings.merge(books,on='ISBN')
```

```python
ratings_with_name.shape
```

```python
num_rating_df=ratings_with_name.groupby('Book-Title')['Book-Rating'].count().reset_index()
```

```python
num_rating_df=num_rating_df.rename(columns={'Book-Rating':'num_ratings'})
```

```python
num_rating_df.head()
```

```python
avg_rating_df=ratings_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_rating_df=avg_rating_df.rename(columns={'Book-Rating':'avg_ratings'})
avg_rating_df.head()
```

```python
popularity_df=num_rating_df.merge(avg_rating_df,on='Book-Title')
popularity_df.head()
```

```python
popularity_df.shape
```

```python
popularity_df=popularity_df[popularity_df['num_ratings']>=250].sort_values(by='avg_ratings',ascending=False).head(50)
```

```python
popularity_df.head(5)
```

```python
popularity_df=popularity_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_ratings']]
```

```python
popularity_df.head()
```

### Collaborative-filtering

```python
ratings_with_name.head(2)
```

```python
x=ratings_with_name.groupby('User-ID')['Book-Rating'].count()>200
experienced_user_id=x[x].index
```

```python
filtered_rating=ratings_with_name[ratings_with_name['User-ID'].isin(experienced_user_id)]
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating',fill_value=0)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
def recommend(book_name):
    # index fetch
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:5]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    
    return data
```

```python
recommend('1984')
```
